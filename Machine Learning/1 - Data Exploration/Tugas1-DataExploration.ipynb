```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./input/european-football-database-20192020/D1.csv')
df = df.iloc[:,[3,4,5,6,11,12,13,14,15,16,19,20,21,22]]
df = df.rename(columns={'FTHG': 'H_Goals', 
                       'FTAG': 'A_Goals', 
                       'HS': 'H_Shots', 
                       'AS': 'A_Shots',
                       'HST': 'H_ShotsOnTarget',
                       'AST': 'A_ShotsOnTarget',
                       'HF': 'H_Fouls', 
                       'AF': 'A_Fouls',
                       'HY': 'H_YellowCards',
                       'AY': 'A_YellowCards',
                       'HR': 'H_RedCards',
                       'AR': 'A_RedCards'})
df
```

```python
meanHomeTeamGoals = df['H_Goals'].mean()
meanHomeTeamGoals
```

```python
medianHomeTeamGoals = df['H_Goals'].median()
medianHomeTeamGoals
```

```python
modusHomeTeamGoals = df['H_Goals'].mode()
modusHomeTeamGoals
```

```python
varianHomeTeamGoals = df['H_Goals'].var()
varianHomeTeamGoals
```

```python
stdHomeTeamGoals = df['H_Goals'].std()
stdHomeTeamGoals
```

```python
plt.title('Distribution of Home Team Goals')
plt.xlabel('Home Team Goals')
plt.ylabel('Frequency')
h = sns.histplot(data=df, x='H_Goals', discrete=True, kde=True)
```

```python
plt.title('Distribution of Home Team Shots')
plt.xlabel('Home Team Shots')
plt.ylabel('Frequency')
h = sns.histplot(data=df, x='H_Shots', discrete=True, kde=True)
```

```python
plt.title('Distribution of Home Team Shots On Target')
plt.xlabel('Home Team Shots OnTarget')
plt.ylabel('Frequency')
h = sns.histplot(data=df, x='H_ShotsOnTarget', discrete=True, kde=True)
```
